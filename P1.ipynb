```python
# An array to save all the links
link_elements = []

# Iterate over each div container
# First of all we make the links global because the links are internal
# Second make a request to all the links and extract the html file of the ads
# Third from that html file we extract the name of the ads and the car informations

# Extract the ads links
for url in urls:
    response = requests.get(url)

    # Parse the HTML content --> this is the html of this domain
    soup = BeautifulSoup(response.content, "html.parser")


    # Find and extract the (car ads) from the html of previos code
    div_containers = soup.find_all("div", class_= 'post-card-item-af972 kt-col-6-bee95 kt-col-xxl-4-e9d46')

    # Iterate over each div container and extract the link of the car ads
    for div_container in div_containers:

        # Find and extract links ('a') from the div container
        new_links = div_container.find_all("a")
        for link in new_links:
            link_elements.append(link)

# Printing the extracted links
for link_element in link_elements:
    print(link_element['href'])
```

```python
# An array to save all the ads
saved_ads = []

# Iterate over each link element and make the request and take the ads informations
for link_element in link_elements:

    # Avariable to save one ads to it like dict
    dict_item = {}

    # Extract the URL of the link and Correction of the links
    link_url = link_element["href"]
    link_url = 'https://divar.ir' + link_url

    # Send a request to the linked ads
    linked_page_response = requests.get(link_url)

    # Parse the HTML content of the linked page (now we have the ads)
    linked_page_soup = BeautifulSoup(linked_page_response.content, "html.parser")

    # Find and extract ads title from the page
    name_data_element = linked_page_soup.find("div", class_="kt-page-title__title kt-page-title__title--responsive-sized")

    # Find and extract ads info from the page
    info = linked_page_soup.find("div", class_="post-page__section--padded")


    # Check if the element exists before extracting text
    if name_data_element and info:

        # Extract the text ads title
        name_text = name_data_element.text.strip()
        print("عنوان آگهی : ", name_text)
        dict_item['عنوان آگهی'] = name_text

        # Find and extract ads information from section
        # The info : color , model , km trsveled
        detail_info_1 = info.find_all("div", class_="kt-group-row-item kt-group-row-item--info-row")

        # The info : other listed informations
        detail_info_2 = info.find_all("div", class_="kt-base-row kt-base-row--large kt-unexpandable-row")

        # Print the pair of info title and value
        if detail_info_2 and detail_info_1:

            # Extract the detail_info_1
            for item in detail_info_1:
                title = item.find("span", class_="kt-group-row-item__title kt-body kt-body--sm")
                val = item.find("span", class_="kt-group-row-item__value")

                # Extract the text and print it
                if title and val:
                    # Extract test
                    title_text = title.text.strip()
                    val_text = val.text.strip()

                    # Print the result
                    print(f'{title_text} : {val_text}')

                    # Save the result
                    dict_item[title_text] = val_text

            # Extract the detail_info_2
            for item in detail_info_2:
                title = item.find("p", class_="kt-base-row__title kt-unexpandable-row__title")
                val = item.find("p", class_="kt-unexpandable-row__value")

                # Extract the text and print it
                if title and val:
                    # Extract test
                    title_text = title.text.strip()
                    val_text = val.text.strip()

                    # Print the result
                    print(f'{title_text} : {val_text}')

                    # Save the result
                    dict_item[title_text] = val_text

        # seprator
        print("-----------------------------------")

        #save the dicts
        saved_ads.append(dict_item)
```

```python
# A method for remove unwanted words
def modify_list_of_dicts(list_of_dicts, substring_to_remove, substring_to_replace):
    modified_list = []

    for dictionary in list_of_dicts:
        modified_dict = {}

        for key, value in dictionary.items():
            modified_key = key.replace(substring_to_remove, substring_to_replace)
            modified_value = value.replace(substring_to_remove, substring_to_replace)

            modified_dict[modified_key] = modified_value

        modified_list.append(modified_dict)

    return modified_list
```

```python
# A method for translate numbers
def trans_num_in_dicts(list_of_dicts):

    translation_dict = {
        '۰': '0','۱': '1','۲': '2','۳': '3','۴': '4',
        '۵': '5','۶': '6','۷': '7','۸': '8','۹': '9'
    }

    modified_list = []

    for dictionary in list_of_dicts:
        modified_dict = {}

        for key, value in dictionary.items():

            translated_items = []

            if isinstance(value, str):
                translated_item = ''.join(translation_dict.get(digit, digit) for digit in value)
                translated_items.append(translated_item)

                if(translated_item == ''):
                    translated_items.append(value)
            else:
                translated_items.append(value)

            modified_dict[key] = translated_items[0]

        modified_list.append(modified_dict)

    return modified_list
```

```python
# Remove unwanted words
saved_ads = modify_list_of_dicts(saved_ads, "\u200c",' ')
saved_ads = modify_list_of_dicts(saved_ads, " تومان", '')
saved_ads = modify_list_of_dicts(saved_ads, "٬", '')
saved_ads = modify_list_of_dicts(saved_ads, " ماه", '')

# Translate numbers
saved_ads = trans_num_in_dicts(saved_ads)

for i in saved_ads:
    print(i)
```

```python
# A method for translating the key and vlaues
def translate_info(list_of_dicts, source_lang, target_lang):
    translator = Translator(service_urls=['translate.google.com'])

    translated_list = []

    for dictionary in list_of_dicts:
        translated_dict = {}

        for key, value in dictionary.items():
            translated_key = translator.translate(key, src=source_lang, dest=target_lang).text
            translated_value = translator.translate(value, src=source_lang, dest=target_lang).text
            translated_dict[translated_key] = translated_value

        translated_list.append(translated_dict)

    return translated_list
```

```python
# Translate thr keys and vaalues
saved_ads_trans = translate_info(saved_ads, 'fa', 'en')

for i in saved_ads_trans:
    print(i)
```

```python
# A method for saving the scarped ads data into a dataset csv file
def save_to_csv(data_list, filename):
    fieldnames = set().union(*(d.keys() for d in data_list))

    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data_list)
```

```python
# Read generated CSV file
df = pd.read_csv('scraped_divar_data.csv')

df.head(5)
```

```python
# Print the shape of the data
print(f"Shape of the dataset: {df.shape}")

# List the columns in the dataset
print(f"Columns in the dataset: {df.columns}")
```

```python
# Preprocessing the dataset
# Step1: Delete irrelevant columns from the dataset

# Identify and delete irrelevant columns
irrelevant_columns = ['Minimum pre -paid amount', 'The ad', 'Number of installments',
                      'The amount of each installment', 'Willing to swap','Ad title']

df = df.drop(irrelevant_columns, axis=1)
df.head(5)
```

```python
# Step2: Handling missing values in a dataset

# Fill missing values in the "Front chassis" column with "healthy"
df['Front chassis'].fillna('Healthy and seal', inplace=True)

# Fill missing values in the "gearbox" column with "manual"
df['gearbox'].fillna('manual', inplace=True)

# Fill missing values in the "Third -party insurance deadline" column with 0
df['Third -party insurance deadline'].fillna(0, inplace=True)

# Fill missing values in the "Engine status" column with 'Healthy'
df['Engine status'].fillna('Healthy', inplace=True)

# Fill missing values in the "The state of chassis" column with 'Healthy and seal'
df['The state of chassis'].fillna('Healthy and seal', inplace=True)

# Fill missing values in the "The state of chassis" column with 'Healthy and seal'
df['The rear chassis'].fillna('Healthy and seal', inplace=True)

df.head(5)
```

```python
# Define a function to convert AD date to solar date
def convert_to_mill(price):
    return price //1000000
```

```python
# Define a function to convert AD date to solar date
def convert_to_solar(year):
    # Check if year is in AD date format
    if year>1800:
        # Conversion formula from AD to solar
        solar_year = year - 621
    else:
        solar_year = year

    return solar_year
```

```python
# Step3: Modify incorrect values

# Apply the conversion function to the "year" column
df['Model (year of production)'] = df['Model (year of production)'].apply(convert_to_solar)

# Apply the conversion function to the "price" column
df['base price'] = df['base price'].apply(convert_to_mill)

df.head(5)
```

```python
# Step4: Handling categorical variables
categorical_columns = ['Front chassis', 'gearbox', 'Color', 'Engine status',
                       'The state of chassis', 'The rear chassis', 'fuel type','Body condition']

encoder = LabelEncoder()
for i in categorical_columns:
    df[i] = encoder.fit_transform(df[i])

df.head(5)
```

```python
# Feature scaling/Normalization
'''numeric_columns = ['base price', 'Third -party insurance deadline' ,
                   'Model (year of production)','Function']  # Replace with the actual numeric column names
scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

df.head(5)'''
```

```python
# Identify and delete irrelevant columns
irrelevant_columns = ['Front chassis', 'gearbox', 'fuel type', 'The rear chassis']

df = df.drop(irrelevant_columns, axis=1)
df.head(5)
```

```python
# Splitting the dataset
X = df.drop('base price', axis=1)
y = df['base price']

# Perform feature scaling using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Printing the shape of the sets
print(f'Shape of X_train: {X_train.shape}')
print(f'Shape of y_train: {y_train.shape}')
print(f'Shape of X_test: {X_test.shape}')
print(f'Shape of y_test: {y_test.shape}')
```

```python
# Train the Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=1000, random_state=42)
rf_regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_regressor.predict(X_test)

print(y_pred)
```

```python
# Calculate evaluation metrics for regression
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print (f'Mean Squared Error (MSE): {mse}')
print (f'Mean Absolute Error (MAE): {mae}')
print (f'Root Mean Squared Error (RMSE): {rmse}')
print (f'Coefficient of Determination (R-squared): {r2}')
```

```python
# Train the Support Vector Regression model
svm_regressor = SVR(kernel='rbf')  # You can experiment with different kernels (e.g., 'linear', 'poly', 'sigmoid')
svm_regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_regressor.predict(X_test)

print(y_pred)
```

```python
# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print (f'Mean Squared Error (MSE): {mse}')
print (f'Mean Absolute Error (MAE): {mae}')
print (f'Root Mean Squared Error (RMSE): {rmse}')
print (f'Coefficient of Determination (R-squared): {r2}')
```
